## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

This lane (Ranking Signal Analysis) is best framed as a **ranking** task.

**Why?** The core objective is to understand which content and search signals are associated with visibility, clicks, engagement, or movement, ultimately influencing how content is ordered or presented to users. This directly aligns with the goal of a ranking system: to arrange items (content) in a sequence based on their estimated relevance or performance.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

The model would predict a **proxy for content quality/relevance** that directly influences its rank. Given the context, a suitable proxy would be **'total_engagement_score'**, which is a composite score derived from observed outcomes like clicks, watch time, shares, and other user interactions. This score would represent how well a piece of content is performing in terms of user interest and interaction. It's an observed outcome, calculated from the `fact_content_daily_performance` data.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

A key success metric for this ranking task would be the **Normalized Discounted Cumulative Gain (NDCG) @ k**. This metric is well-suited for ranking problems as it considers the graded relevance of items and their position in a ranked list. It evaluates the quality of the ranking by penalizing highly relevant items if they appear lower in the list.

**Why NDCG?** It is a robust metric that accounts for the position of relevant items and the varying degrees of relevance (e.g., high engagement is more relevant than low engagement). A higher NDCG @ k means that the model is effectively surfacing the most engaging content at the top of the ranked list, which translates directly to the business goal of increasing visibility and user interaction with quality content.

## 4. The unit of analysis, as a real dataframe

The unit of analysis for this problem would be **one content item on a given day**. Each row in the dataset represents the daily performance and associated signals for a unique piece of content. ## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

ML beats a fixed rule here because the relationship between content signals and actual user engagement/ranking performance is highly **complex, non-linear, and dynamic**.

1.  **Multiple Interacting Signals**: Content performance is not driven by one or two signals, but by a combination of many (e.g., content type, search signals, time of day, user characteristics, external trends). Fixed rules struggle to account for the intricate interactions and varying importance of these signals.
2.  **Evolving Patterns**: User behavior, content trends, and search algorithms constantly change. A fixed rule would quickly become outdated and require constant manual updates. ML models can adapt and learn from new data, capturing these evolving patterns.
3.  **Non-linear Relationships**: The impact of a signal on engagement might not be linear. For example, a certain signal might be beneficial up to a point, but detrimental thereafter. ML models can uncover these non-linear relationships that are impossible to define with simple if-else statements.
4.  **Scalability**: Manually crafting rules for millions of content pieces and billions of user interactions is infeasible. ML provides a scalable way to learn these patterns automatically from data.
